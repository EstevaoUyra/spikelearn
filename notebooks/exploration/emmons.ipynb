```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
t = np.linspace(0,12,100)
pc1 = -np.sin(t/2)
pc2 = st.norm(6, 1.8).pdf(t)
```

```python
plt.figure(figsize=(8,6))
plt.subplot(2,2,1)
plt.plot(t,pc1)

plt.subplot(2,2,3)
plt.plot(t,pc2)

plt.subplot(1,2,2)
plt.plot(pc2, pc1)
plt.tight_layout()
```
