```python
import os
os.chdir('../../')
from spikelearn.data.preprocessing import kernel_smooth
from spikelearn.data import io, SHORTCUTS
import pandas as pd

BASELINE = -500
DSET_PARAMS = {'wide_smoothed' : { 'sigma' : 100,
                                  'bin_size' : 100},

              'narrow_smoothed' : { 'sigma' : 50,
                                    'bin_size' : 100},

              'wide_smoothed_viz' : { 'sigma' : 100,
                                      'bin_size' : 20},

              'narrow_smoothed_viz' : { 'sigma' : 50,
                                        'bin_size' : 20}
                }
```

```python
rat_label = 'DRRD 7'
epoched = io.load(rat_label, 'epoched_spikes')
epoched = epoched.groupby('is_selected').get_group(True)
```

```python
ep = epoched.reset_index().set_index(['trial','unit'])
```

```python
smoothed_dataset = pd.DataFrame(index = ep.index)
```

```python
smoothed_dataset = smoothed_dataset.reset_index().set_index(['trial','unit'])
```

```python
dset_name = 'wide_smoothed'
params = DSET_PARAMS[dset_name]
```

```python
f = lambda x: kernel_smooth( 1000*x['with_baseline'], **params,
                                      edges=(BASELINE, 1000*x.duration))
out = ep.reset_index().iloc[::40].apply(f, axis=1)
```
